```python
#Q19. What is the relationship between a Z-score and a confidence interval?
"""
Ans19. The relationship between Z-score and a confidence interval lies in how the Z-score determines
       the range of the confidence interval.

       1. Z-Score as a Multiplier:
                   The Z-score tells us how many standard errors we need to go and below the sample
                   estimate(likley the sample mean) to create a confidence interval.
                   for example: For 95% confidence, the Z-score is 1.96. this means the confidence 
                   interval will stretch 1.96 times the standard error in both directions.
        
        2. Confidence Interval Formula:
                    The confidence interval is calculoated as:
                    confidence interval = Sample estimate +-(Z-score * Standard Error)
            The Z-score defines how far the inerval extends from the sample estimate.
        
        3. Higher Confidence Levels:
                    A higher confidence level(e.g.,99%) requires a larger Z-score, resulting
                    in a wider confidence interval to account fo greater uncertainity.
                    
"""
```

```python
#Q22.  What is the concept of expected value in a probability distribution?
"""
Ans22. Expected value in a probability distribution is a mean of a random variable.
        In simple terms:
        . Multiply each possible value by its probability.
        . Add all those products together.
        . The result is the expected value.
        For example: If you roll a dice, the possible outcomes are 1,2,3,4,5,6 and each equal
                    probability of 1/6
        the Expected value is : (1*1/6)+(2*1/6)+(3*1/6)+(4*1/6)+(5*1/6)+(6*1/6)
                             = 3.5
                This means that, on average, if you roll the dice many times, the result will be
                3.5, even though you can never roll a 3.5 on a single roll. The expected value
                shows the "long term average" outcome when the experiment is repeated multiple times.
"""
```

```python
#Q19. Implement hypothesis testing using Z-statistics for a sample dataset.
"""
Scenario: Suppose you are a coffee shop owner, and you want to test whether the average time customers spend
          in your coffee shop has increased.

          1. Population detail
              . From past data, you know that the average time a customer spends in your shop is 45 minutes
                (population mean, ùúá).
              . The standard deviation of customer time is 5 minutes (population standard deviation, ùúé).

          2. Sample detail
              .  You take a random sample of 30 customers and find that the sample mean is 47 minutes.

            ùúá = 45,  ùúé=5, n =30, sample mean= 47 

1. Frame the hypothesis
        Null Hypothesis(H0) = ùúá = 45
        Alternate hypothesis(HA) = ùúá > 45

2. Significance level = 0.05
"""
```

```python
population_mean = 45 
population_std = 5
sample_mean = 47
sample_size = 30
alpha = 0.05

z_statistic = (sample_mean-population_mean)/(population_std/np.sqrt(sample_size))
```

```python
z_statistic
```

```python
p_value = 1-norm.cdf(z_statistic)
```

```python
p_value
```

```python
if p_value < alpha:
    print("reject the null hypothesis")
else:
    print("failed to reject the  null hypothesis")
```

```python
#Q28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

from scipy.stats import skew, kurtosis

def skew_kurtosis(mean,std,size):
    data = np.random.normal(mean,std,size)

    data_skewness = skew(data)
    data_kurtosis = kurtosis(data, fisher = True)

    print(f'Skewness of a dataset: {data_skewness:.4f}')
    print(f'Skewness of a dataset:{data_kurtosis:.4f}')

    return data, data_skewness, data_kurtosis
    
    
```
